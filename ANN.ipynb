```python
from keras.models import Sequential
from keras.layers import Dense         # this is used to define the number of hidden layers
import numpy
```

```python
seed = 7
numpy.random.seed(seed)

data = numpy.loadtxt("pima-indians-diabetes.data.csv",delimiter = ',')
```

```python
x = data[:,0:8]
y = data[:,8]
```

#### Building a framework for neural network

```python
# The neural network consist of different layers and we want to build those in a sequence hence we will call the function sequential

model = Sequential()
model.add(Dense(12,input_dim = 8,kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(8,kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(1,kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
from IPython import display
display.Image("G:/image1.png")
```

#### Compiling the model

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.fit(x,y, validation_split = 0.33, epochs = 150, batch_size = 10)
```

```python
scores = model.evaluate(x,y)
print("%s: %.2f%%" % (model.metrics_names[1],scores[1]*100))
```
